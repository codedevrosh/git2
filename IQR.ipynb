```python

col=['Height', 'Weight']
for c in col:
  Q1=df[c].quantile(0.25)
  Q3=df[c].quantile(0.75)
  IQR=Q3-Q1
  upper_limit=Q3+(IQR*1.5)
  lower_limit=Q1-(IQR*1.5)
  print(f"{c}->{lower_limit},{upper_limit}")
  outliers = df[(df[c] < lower_limit) | (df[c] > upper_limit)]
  print(f"Outliers in {c}:")
  print(outliers, "\n")
  df = df[(df[c] >= lower_limit) & (df[c] <= upper_limit)]
df
```
